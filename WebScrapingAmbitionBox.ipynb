```python

import requests
from bs4 import BeautifulSoup

url = "https://www.ambitionbox.com/list-of-companies?page=1"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all company name elements
    company_names = soup.find_all("h2", class_="companyCardWrapper__companyName")

    # Find all review count elements
    review_counts = soup.find_all("span", class_="companyCardWrapper__ActionCount")

    # Iterate over the company names and review counts and print them
    for name, count in zip(company_names, review_counts):
        company_name = name.text.strip()
        review_count = count.text.strip()
        print(company_name, review_count)
else:
    print("Failed to retrieve data from the website.")
```
